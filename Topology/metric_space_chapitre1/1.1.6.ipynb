```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Explication de voisinage d'un point dans un espace métrique, en se concentrant sur la distinction entre un point intérieur et un point frontière. Le code génère une visualisation claire pour illustrer pourquoi un intervalle ouvert autour d'un point intérieur reste entièrement dans l'ensemble, tandis qu'un intervalle autour d'un point frontière déborde hors de l'ensemble.
on a l'ensemble A = [1, 2[ sur la droite réelle. On examine deux points :
- x = 1.5 (intérieur à A)
- x = 1 (frontière de A)
- Pour x = 1.5, on peut trouver un intervalle ouvert (voisinage) qui reste entièrement dans A.
- Pour x = 1, tout intervalle ouvert autour de ce point déborde hors de A.
- La visualisation utilise des couleurs et des annotations pour clarifier ces concepts.

```python
# Configuration du style
sns.set_theme(style="white")

def visualiser_voisinage_frontiere():
    fig, ax = plt.subplots(figsize=(12, 6))

    # 1. Dessiner l'axe des réels
    ax.axhline(y=0, color='black', linewidth=1, zorder=1)

    # 2. Dessiner l'ensemble A = [1, 2[
    # On utilise une ligne épaisse bleue
    # Le crochet fermé [ en 1 et ouvert [ en 2 sont simulés par les marqueurs
    ax.plot([1, 2], [0, 0], color='#3498db', linewidth=8, label='Ensemble A = [1, 2[', zorder=2)

    # Marqueur pour 1 (Fermé / Inclus)
    ax.scatter(1, 0, color='#3498db', s=200, marker='|', linewidth=4, zorder=3)
    # Marqueur pour 2 (Ouvert / Exclu) - un cercle vide
    ax.scatter(2, 0, color='white', edgecolor='#3498db', s=150, linewidth=3, zorder=3)

    # ---------------------------------------------------------
    # CAS 1 : Point Intérieur x = 1.5
    # ---------------------------------------------------------
    x_int = 1.5
    ax.scatter(x_int, 0, color='black', s=100, label='Point x = 1.5 (Intérieur)', zorder=4)

    # Dessiner un voisinage valide (Ouvert inclus)
    epsilon_int = 0.2
    ax.plot([x_int - epsilon_int, x_int + epsilon_int], [0, 0],
            color='#2ecc71', linewidth=4, alpha=0.8,
            label='Voisinage Ouvert valide (Inclus)')

    # Annotation pour le succès
    ax.annotate("SUCCÈS : L'intervalle vert\nest entièrement dans A",
                xy=(1.5, 0.02), xytext=(1.5, 0.2),
                arrowprops=dict(facecolor='#2ecc71', shrink=0.05),
                ha='center', color='#27ae60', fontweight='bold')

    # ---------------------------------------------------------
    # CAS 2 : Point Frontière x = 1
    # ---------------------------------------------------------
    x_bord = 1.0
    ax.scatter(x_bord, 0, color='red', s=100, label='Point x = 1 (Frontière)', zorder=4)

    # Dessiner une tentative de voisinage (qui déborde)
    epsilon_bord = 0.15
    # Partie qui est DANS l'ensemble
    ax.plot([1, 1 + epsilon_bord], [0, 0], color='#e74c3c', linewidth=4, alpha=0.5)
    # Partie qui DÉBORDE (HORS de l'ensemble)
    ax.plot([1 - epsilon_bord, 1], [0, 0], color='#e74c3c', linewidth=4, linestyle='--',
            label='Tentative d\'ouvert (Déborde)')

    # Annotation pour l'échec
    ax.annotate("ÉCHEC : Ça déborde ici !\n(0.9, 0.99... ne sont pas dans A)",
                xy=(0.92, 0), xytext=(0.5, 0.2),
                arrowprops=dict(facecolor='#e74c3c', shrink=0.05),
                ha='center', color='#c0392b', fontweight='bold')

    # Mise en forme
    ax.set_xlim(0.4, 2.6)
    ax.set_ylim(-0.1, 0.4)
    ax.set_yticks([]) # Cacher l'axe Y
    ax.set_xticks([1, 1.5, 2])
    ax.set_xticklabels(['1\n(Fermé)', '1.5', '2\n(Ouvert)'], fontsize=12)

    ax.legend(loc='upper right')
    ax.set_title("Visualisation : Pourquoi [1, 2[ est voisinage de 1.5 MAIS PAS de 1", fontsize=16, pad=20)

    # Enlever les cadres inutiles
    sns.despine(left=True, bottom=False)

    plt.tight_layout()
    plt.show()

visualiser_voisinage_frontiere()
```
