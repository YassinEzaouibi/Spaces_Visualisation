```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_theme(style="white")


def demo_densite():
    fig, ax = plt.subplots(figsize=(10, 4))

    # Masquer les axes Y
    ax.get_yaxis().set_visible(False)
    ax.set_ylim(-1, 1)
    ax.set_xlim(0, 3)

    # 1. LA CIBLE (Un Ouvert)
    # Imaginons qu'on cherche à atteindre un nombre entre 1.4 et 1.6
    ax.axvspan(1.4, 1.6, color='#e74c3c', alpha=0.3, label='Ouvert Cible ]1.4, 1.6[')

    # 2. ENSEMBLE NON DENSE (Les Entiers Z)
    # Points : 0, 1, 2, 3...
    integers = [0, 1, 2, 3]
    ax.scatter(integers, [0.2] * len(integers), color='green', s=100, marker='s',
               label='Entiers $\mathbb{Z}$ (Non Dense)')

    # 3. ENSEMBLE DENSE (Les Rationnels Q)
    # On génère plein de fractions k/n
    rationals = []
    for den in range(1, 20):  # Dénominateurs
        for num in range(0, 3 * den):  # Numérateurs
            rationals.append(num / den)

    # On en affiche un échantillon
    ax.scatter(rationals, [-0.2] * len(rationals), color='blue', s=20, alpha=0.6,
               label='Rationnels $\mathbb{Q}$ (Dense)')

    # Analyse du résultat
    ax.annotate("Z rate la cible !\n(Il y a un trou)",
                xy=(1.5, 0.2), xytext=(1.5, 0.6),
                arrowprops=dict(facecolor='green', arrowstyle='->'), ha='center', color='green')

    ax.annotate("Q touche la cible !\n(Il y a des fractions dedans)",
                xy=(1.5, -0.2), xytext=(1.5, -0.6),
                arrowprops=dict(facecolor='blue', arrowstyle='->'), ha='center', color='blue')

    plt.title("Visualisation : Dense vs Non-Dense", fontsize=14)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()


demo_densite()
```
