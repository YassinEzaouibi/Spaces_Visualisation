```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style="whitegrid")

def visualiser_continuite():
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # -------------------------------------------------------
    # 1. Fonction CONTINUE (La Sigmoïde)
    # -------------------------------------------------------
    x = np.linspace(-6, 6, 200)
    # Formule sigmoïde : 1 / (1 + e^-x)
    y_cont = 1 / (1 + np.exp(-x))

    axes[0].plot(x, y_cont, color='#2ecc71', linewidth=4, label='Sigmoïde')
    axes[0].set_title("Fonction CONTINUE\n(Pas de saut, le crayon reste sur la feuille)", fontsize=14)

    # Annotation
    axes[0].annotate("Transition douce", xy=(0, 0.5), xytext=(-4, 0.8),
                     arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12)

    # -------------------------------------------------------
    # 2. Fonction DISCONTINUE (Heaviside / Step)
    # -------------------------------------------------------
    # Partie gauche
    x_left = np.linspace(-6, 0, 100)
    y_left = np.zeros_like(x_left) # 0 partout

    # Partie droite
    x_right = np.linspace(0, 6, 100)
    y_right = np.ones_like(x_right) # 1 partout

    axes[1].plot(x_left, y_left, color='#e74c3c', linewidth=4)
    axes[1].plot(x_right, y_right, color='#e74c3c', linewidth=4, label='Heaviside')

    # Le SAUT (ligne pointillée verticale)
    axes[1].plot([0, 0], [0, 1], color='black', linestyle='--', alpha=0.5, label='Saut (Discontinuité)')

    # Les points ouverts/fermés pour la rigueur
    axes[1].scatter([0], [0], s=150, facecolors='white', edgecolors='#e74c3c', linewidth=2, zorder=5) # Exclu en bas
    axes[1].scatter([0], [1], s=150, color='#e74c3c', zorder=5) # Inclus en haut

    axes[1].set_title("Fonction DISCONTINUE en 0\n(Saut brutal, limite gauche != limite droite)", fontsize=14)

    # Annotation du danger
    axes[1].annotate("RUPTURE !\nIci la limite explose", xy=(0, 0.5), xytext=(2, 0.5),
                     arrowprops=dict(facecolor='#c0392b', shrink=0.05), color='#c0392b', fontweight='bold')

    plt.tight_layout()
    plt.show()

visualiser_continuite()
```
