```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# Set up the figure and subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Bounded Above & Below Visualizations', fontsize=16, fontweight='bold')

# 1. Bounded Above Example: y = -x²
x1 = np.linspace(-3, 3, 100)
y1 = -x1**2
ax1.plot(x1, y1, 'b-', linewidth=2, label='y = -x²')
ax1.axhline(y=0, color='r', linestyle='--', linewidth=2, label='Upper Bound = 0')
ax1.fill_between(x1, y1, -10, alpha=0.3, color='blue')
ax1.set_title('Bounded Above\n(Has a "ceiling")')
ax1.set_ylabel('y-values')
ax1.set_ylim(-10, 1)
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Bounded Below Example: y = x²
x2 = np.linspace(-3, 3, 100)
y2 = x2**2
ax2.plot(x2, y2, 'g-', linewidth=2, label='y = x²')
ax2.axhline(y=0, color='r', linestyle='--', linewidth=2, label='Lower Bound = 0')
ax2.fill_between(x2, y2, 10, alpha=0.3, color='green')
ax2.set_title('Bounded Below\n(Has a "floor")')
ax2.set_ylabel('y-values')
ax2.set_ylim(-1, 10)
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Bounded Example: y = sin(x)
x3 = np.linspace(-2*np.pi, 2*np.pi, 200)
y3 = np.sin(x3)
ax3.plot(x3, y3, 'purple', linewidth=2, label='y = sin(x)')
ax3.axhline(y=1, color='r', linestyle='--', linewidth=2, label='Upper Bound = 1')
ax3.axhline(y=-1, color='orange', linestyle='--', linewidth=2, label='Lower Bound = -1')
ax3.fill_between(x3, y3, -1, alpha=0.3, color='purple')
ax3.set_title('Bounded (Both Above & Below)\n(In a "corral")')
ax3.set_xlabel('x')
ax3.set_ylabel('y-values')
ax3.set_ylim(-1.5, 1.5)
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Unbounded Example: y = x
x4 = np.linspace(-5, 5, 100)
y4 = x4
ax4.plot(x4, y4, 'red', linewidth=2, label='y = x')
ax4.set_title('Unbounded\n(No upper or lower bounds)')
ax4.set_xlabel('x')
ax4.set_ylabel('y-values')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Additional visualization: Discrete sets with bounds
fig2, (ax5, ax6, ax7) = plt.subplots(1, 3, figsize=(15, 5))

# Bounded Above Discrete Set
set_a = [-3, -2, -1, 0, 1]
ax5.scatter(set_a, [0]*len(set_a), color='blue', s=100, zorder=3)
ax5.axhline(y=0, color='red', linestyle='--', linewidth=3, label='Upper Bound = 1')
ax5.axhline(y=0, color='pink', linestyle=':', linewidth=2, label='Other Upper Bounds (≥1)')
for i, val in enumerate(set_a):
    ax5.text(val, 0.1, str(val), ha='center', va='bottom')
ax5.set_title('Discrete Set: Bounded Above\n{..., -3, -2, -1, 0, 1}')
ax5.set_xlim(-4, 3)
ax5.set_ylim(-0.5, 1)
ax5.legend()

# Bounded Below Discrete Set
set_b = [1, 2, 3, 4, 5]
ax6.scatter(set_b, [0]*len(set_b), color='green', s=100, zorder=3)
ax6.axhline(y=0, color='orange', linestyle='--', linewidth=3, label='Lower Bound = 1')
ax6.axhline(y=0, color='lightgreen', linestyle=':', linewidth=2, label='Other Lower Bounds (≤1)')
for i, val in enumerate(set_b):
    ax6.text(val, 0.1, str(val), ha='center', va='bottom')
ax6.set_title('Discrete Set: Bounded Below\n{1, 2, 3, 4, 5, ...}')
ax6.set_xlim(0, 6)
ax6.set_ylim(-0.5, 1)
ax6.legend()

# Bounded Discrete Set
set_c = [1, 1/2, 1/3, 1/4, 1/5]
ax7.scatter(set_c, [0]*len(set_c), color='purple', s=100, zorder=3)
ax7.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Upper Bound = 1')
ax7.axhline(y=0, color='orange', linestyle='--', linewidth=2, label='Lower Bound = 0')
for i, val in enumerate(set_c):
    ax7.text(val, 0.1, f'{val:.2f}', ha='center', va='bottom')
ax7.set_title('Discrete Set: Bounded\n{1, 1/2, 1/3, 1/4, 1/5, ...}')
ax7.set_xlim(-0.1, 1.2)
ax7.set_ylim(-0.5, 1)
ax7.legend()

plt.tight_layout()
plt.show()

# Interactive visualization to explore bounds
def plot_with_bounds(function_str, x_range=(-5, 5), upper_bound=None, lower_bound=None):
    """Plot a function with optional upper and lower bounds"""
    fig, ax = plt.subplots(figsize=(10, 6))

    x = np.linspace(x_range[0], x_range[1], 400)

    # Evaluate the function safely
    try:
        y = eval(function_str)
    except:
        print(f"Error evaluating function: {function_str}")
        return

    ax.plot(x, y, 'b-', linewidth=2, label=f'y = {function_str}')

    # Plot bounds if provided
    if upper_bound is not None:
        ax.axhline(y=upper_bound, color='red', linestyle='--',
                  linewidth=2, label=f'Upper Bound = {upper_bound}')
        # Shade the region above the bound
        ax.fill_between(x, upper_bound, max(y) + 1, alpha=0.2, color='red')

    if lower_bound is not None:
        ax.axhline(y=lower_bound, color='green', linestyle='--',
                  linewidth=2, label=f'Lower Bound = {lower_bound}')
        # Shade the region below the bound
        ax.fill_between(x, min(y) - 1, lower_bound, alpha=0.2, color='green')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Function: y = {function_str}\nwith Bounds')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.show()

    # Determine boundedness
    is_bounded_above = upper_bound is not None and all(y <= upper_bound)
    is_bounded_below = lower_bound is not None and all(y >= lower_bound)

    print(f"\nAnalysis for y = {function_str}:")
    print(f"Bounded above: {is_bounded_above}")
    print(f"Bounded below: {is_bounded_below}")
    if is_bounded_above and is_bounded_below:
        print("Status: BOUNDED")
    elif is_bounded_above:
        print("Status: BOUNDED ABOVE")
    elif is_bounded_below:
        print("Status: BOUNDED BELOW")
    else:
        print("Status: UNBOUNDED")

# Example usage of the interactive function
print("Interactive Bounds Explorer:")
plot_with_bounds("-x**2", upper_bound=0)  # Bounded above
plot_with_bounds("x**2", lower_bound=0)   # Bounded below
plot_with_bounds("np.sin(x)", upper_bound=1, lower_bound=-1)  # Bounded
plot_with_bounds("x")  # Unbounded
```

```python
# More examples to try:
plot_with_bounds("np.cos(x)", upper_bound=1, lower_bound=-1)
plot_with_bounds("1/(1 + np.exp(-x))", upper_bound=1, lower_bound=0)  # Sigmoid
plot_with_bounds("np.arctan(x)", upper_bound=np.pi/2, lower_bound=-np.pi/2)
```
