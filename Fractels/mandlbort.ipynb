```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandelbrot(width, height, xmin, xmax, ymin, ymax, max_iter):
    """
    Generates the Mandelbrot set for a given window and resolution.
    """
    # Create arrays for the real and imaginary parts of the complex plane
    r = np.linspace(xmin, xmax, width)
    i = np.linspace(ymin, ymax, height)

    # Create a grid of complex numbers
    c = r[np.newaxis, :] + i[:, np.newaxis] * 1j

    # Initialize z and the fractal image
    z = np.zeros_like(c, dtype=np.complex128)
    fractal = np.zeros(c.shape, dtype=int)

    # Iterate to determine which points are in the set
    for n in range(max_iter):
        # Identify points that have not yet diverged
        mask = np.abs(z) <= 2.0

        # Update z for those points
        z[mask] = z[mask]**2 + c[mask]

        # Update the fractal image with the iteration count
        fractal[mask] = n

    return fractal
```

```python
# Image parameters
WIDTH = 800
HEIGHT = 800
MAX_ITER = 100

# Plotting window (complex plane coordinates)
XMIN, XMAX = -2.0, 1.0
YMIN, YMAX = -1.5, 1.5

# Generate the fractal
mandelbrot_set = mandelbrot(WIDTH, HEIGHT, XMIN, XMAX, YMIN, YMAX, MAX_ITER)

# Visualize the fractal
plt.figure(figsize=(10, 10))
plt.imshow(mandelbrot_set, extent=[XMIN, XMAX, YMIN, YMAX], cmap='hot')
plt.colorbar(label='Iterations')
plt.title('Mandelbrot Fractal')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```
